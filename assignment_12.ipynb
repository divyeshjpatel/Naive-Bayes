```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('D:/ExcelR/A12 naive bayes/SalaryData_Train.csv')
```

```python
dataset.head()
```

```python
dataset.drop(columns='education',inplace=True)
dataset['Salary'].replace({' >50K':1,' <=50K':0},inplace=True)
dataset = pd.concat([dataset,pd.get_dummies(dataset['workclass'])],axis=1)
dataset.drop(columns='workclass',inplace=True)
dataset = pd.concat([dataset,pd.get_dummies(dataset['maritalstatus'])],axis=1)
dataset.drop(columns='maritalstatus',inplace=True)
dataset = pd.concat([dataset,pd.get_dummies(dataset['occupation'])],axis=1)
dataset.drop(columns='occupation',inplace=True)
dataset = pd.concat([dataset,pd.get_dummies(dataset['relationship'])],axis=1)
dataset.drop(columns='relationship',inplace=True)
dataset = pd.concat([dataset,pd.get_dummies(dataset['race'])],axis=1)
dataset.drop(columns='race',inplace=True)
dataset = pd.concat([dataset,pd.get_dummies(dataset['sex'])],axis=1)
dataset.drop(columns='sex',inplace=True)
dataset = pd.concat([dataset,pd.get_dummies(dataset['native'])],axis=1)
dataset.drop(columns='native',inplace=True)
titles = list(dataset.columns)
titles[0],titles[5] = titles[5],titles[0]
dataset = dataset[titles]
```

```python
dataset.head()
```

```python
X_train = dataset.iloc[:, 1:].values
y_train = dataset.iloc[:, 0].values
```

```python
test = pd.read_csv('D:/ExcelR/A12 naive bayes/SalaryData_Test.csv')
```

```python
test.head()
```

```python
test.drop(columns='education',inplace=True)
test['Salary'].replace({' >50K':1,' <=50K':0},inplace=True)
test = pd.concat([test,pd.get_dummies(test['workclass'])],axis=1)
test.drop(columns='workclass',inplace=True)
test = pd.concat([test,pd.get_dummies(test['maritalstatus'])],axis=1)
test.drop(columns='maritalstatus',inplace=True)
test = pd.concat([test,pd.get_dummies(test['occupation'])],axis=1)
test.drop(columns='occupation',inplace=True)
test = pd.concat([test,pd.get_dummies(test['relationship'])],axis=1)
test.drop(columns='relationship',inplace=True)
test = pd.concat([test,pd.get_dummies(test['race'])],axis=1)
test.drop(columns='race',inplace=True)
test = pd.concat([test,pd.get_dummies(test['sex'])],axis=1)
test.drop(columns='sex',inplace=True)
test = pd.concat([test,pd.get_dummies(test['native'])],axis=1)
test.drop(columns='native',inplace=True)
cols = list(test.columns)
cols[0],cols[5] = cols[5],cols[0]
test = test[cols]
```

```python
test.head()
```

```python
X_test = test.iloc[:, 1:].values
y_test = test.iloc[:, 0].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
